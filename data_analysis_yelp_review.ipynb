## Identify Common Links

### Subtask:
Find the links that are present in both DataFrames. This will involve inspecting the columns of both datasets to determine the appropriate column(s) for comparison.


**Reasoning**:
To compare the links effectively, I need to standardize them by removing any query parameters. I'll define a function to extract the base URL and then apply it to the 'Yelp Link' column in `df_yelp` and the 'Sample Review' column in `df_ipo` to create new 'base_link' columns.



````python
from urllib.parse import urlparse, urlunparse

def get_base_url(url):
    """Extracts the base URL without query parameters."""
    if pd.isna(url): # Handle potential NaN values
        return None
    parsed_url = urlparse(url)
    return urlunparse(parsed_url._replace(query='', fragment=''))

# Apply the function to df_yelp['Yelp Link']
df_yelp['yelp_base_link'] = df_yelp['Yelp Link'].apply(get_base_url)

# Apply the function to df_ipo['Sample Review']
df_ipo['ipo_base_link'] = df_ipo['Sample Review'].apply(get_base_url)

print("df_yelp with new 'yelp_base_link' column:")
print(df_yelp[['Yelp Link', 'yelp_base_link']].head())
print("\ndf_ipo with new 'ipo_base_link' column:")
print(df_ipo[['Sample Review', 'ipo_base_link']].head())
````

**Reasoning**:
Now that the base URLs have been extracted and standardized, I will find the common links by taking the intersection of the unique 'yelp_base_link' values from `df_yelp` and the unique 'ipo_base_link' values from `df_ipo`. Finally, I will print the count of common links found.



````python
common_links = pd.Series(list(set(df_yelp['yelp_base_link'].dropna()) & set(df_ipo['ipo_base_link'].dropna())))

print(f"Number of common links found: {len(common_links)}")

if not common_links.empty:
    print("\nFirst 5 common links:")
    print(common_links.head())
else:
    print("\nNo common links found.")
````

**Reasoning**:
To standardize the address columns for comparison, I will first define a dictionary for state abbreviations and then create a function to clean and normalize address strings, applying this function to the relevant address columns in both DataFrames while incorporating state abbreviation logic.



````python
import re

# 1. Define the state_abbreviations dictionary
state_abbreviations = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH',
    'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC',
    'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA',
    'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN',
    'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA',
    'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY',
    'District of Columbia': 'DC' # Added DC for completeness
}

# 2. Define the clean_and_standardize_address function
def clean_and_standardize_address(address, full_state_name=None):
    if pd.isna(address):
        return None
    address = str(address).lower()

    # Remove common patterns like suite/apartment numbers, street designators
    address = re.sub(r'suite\s\S+', '', address)
    address = re.sub(r'ste\s\S+', '', address)
    address = re.sub(r'apt\s\S+', '', address)
    address = re.sub(r'#\S+', '', address)
    address = re.sub(r'unit\s\S+', '', address)
    address = re.sub(r'bldg\s\S+', '', address)

    # Replace full street names with abbreviations or remove them if irrelevant
    address = re.sub(r'street', 'st', address)
    address = re.sub(r'avenue', 'ave', address)
    address = re.sub(r'road', 'rd', address)
    address = re.sub(r'boulevard', 'blvd', address)
    address = re.sub(r'drive', 'dr', address)
    address = re.sub(r'lane', 'ln', address)
    address = re.sub(r'parkway', 'pkwy', address)
    address = re.sub(r'court', 'ct', address)
    address = re.sub(r'terrace', 'ter', address)
    address = re.sub(r'place', 'pl', address)
    address = re.sub(r'north', 'n', address)
    address = re.sub(r'south', 's', address)
    address = re.sub(r'east', 'e', address)
    address = re.sub(r'west', 'w', address)

    # Replace full state names with abbreviations using the provided dictionary
    if full_state_name and full_state_name in state_abbreviations:
        address = address.replace(full_state_name.lower(), state_abbreviations[full_state_name].lower())
    else:
        # Also try to replace state names if they appear directly in the address string
        for state, abbr in state_abbreviations.items():
            address = address.replace(state.lower(), abbr.lower())

    # Remove punctuation
    address = re.sub(r'[.,-]', '', address)

    # Normalize whitespace (replace multiple spaces with a single space, strip leading/trailing spaces)
    address = re.sub(r'\s+', ' ', address).strip()

    return address

# 3. Apply the function to df_yelp['Location']
df_yelp['cleaned_address'] = df_yelp['Location'].apply(clean_and_standardize_address)

# 4. Apply the function to df_ipo['Address (Official)'], passing the corresponding 'State' value
df_ipo['cleaned_address'] = df_ipo.apply(lambda row: clean_and_standardize_address(row['Address (Official)'], row['State']), axis=1)

# 5. Print the head of df_yelp
print("\ndf_yelp with 'Location' and 'cleaned_address':")
print(df_yelp[['Location', 'cleaned_address']].head())

# 6. Print the head of df_ipo
print("\ndf_ipo with 'Address (Official)', 'State', and 'cleaned_address':")
print(df_ipo[['Address (Official)', 'State', 'cleaned_address']].head())
````
